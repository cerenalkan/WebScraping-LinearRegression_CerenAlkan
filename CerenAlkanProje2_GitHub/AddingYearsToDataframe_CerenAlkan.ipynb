```python
import numpy as np
```

```python
YEARS = np.empty(1600)
```

```python
YEARS[:200] = 2011
YEARS[200:400] = 2012
YEARS[400:600] = 2013
YEARS[600:800] = 2014
YEARS[800:1000] = 2015
YEARS[1000:1200] = 2016
YEARS[1200:1400] = 2017
YEARS[1400:1600] = 2018

# This is just for performance of the model, we write 1 instead of 2011 here.
YEARS = YEARS - 2010
```

```python
df["Year"] = YEARS
```

```python
df["Year"] = df["Year"].astype(np.int8)
```
