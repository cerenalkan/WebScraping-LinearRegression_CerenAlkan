```python
from urllib.request import urlopen
import requests
from bs4 import BeautifulSoup as bts
import pandas as pd
import numpy as np
import re
from requests_html import HTMLSession
```

```python
def getAndParseURL(url):
    session = HTMLSession()
    result = session.get(url)
    soup = bts(result.text, 'html.parser')
    return soup
```

```python
#We need to scrap the URLs of all movies, so we create list for URLs. 
MOVIE_URL = []

def find_movie_url(date):
    url = ("https://www.boxofficemojo.com/year/%s/?grossesOption=calendarGrosses" %(date))
    html = getAndParseURL(str(url))
               
    for i in html.findAll("td",{"class":"a-text-left mojo-field-type-release mojo-cell-wide"}):        
        MOVIE_URL.append("https://www.boxofficemojo.com/"+i.a["href"])
        
find_movie_url(2017)
print(MOVIE_URL)
```

```python
len(MOVIE_URL)
```

```python
#This section is not neccessary if you don't want to look at the film names.
MOVIE_NAME = []
for pr in MOVIE_URL:                   
    prod = getAndParseURL(pr)
    NAME = prod.find("h1",{"class":"a-size-extra-large"}).text
    MOVIE_NAME.append(NAME)      
         
print(MOVIE_NAME)
```

```python
TT = []
for pr in MOVIE_URL:                   
    prod = getAndParseURL(pr)
    for i in prod.findAll("div",{"class":"a-section a-spacing-none"}):
        for b in i.findAll("a",attrs={'href': re.compile(r'title/tt')}):
            TT.append(str(b["href"]))

print(TT)
```

```python
len(TT)
```

```python
#This section is for building a connection between boxofficemojo and imdb.
TT_list = []
url = []
for x in range(len(TT)):
    
    txt = TT[x]
    txt = txt.replace("https://pro.imdb.com/title/","")
    TT_list.append(txt[0:txt.index("?")])
    #print(txt[0:txt.index("?")])
    url.append("https://www.boxofficemojo.com/title/" + TT_list[x])
    
print(len(url))
```

```python
dt = pd.DataFrame({"Title":TT_list})
dt
```

```python
BUDGETS = np.empty(shape = len(TT_list))
DOMESTIC_DISTRIBUTORS = [np.NaN] * len(TT_list)
MPAAS = [np.NaN] * len(TT_list)
WORLDWIDES = np.empty(shape = len(TT_list))
GENRES = [np.NaN] * len(TT_list)
OPENING = np.empty(shape = len(TT_list))

BUDGETS[:] = np.NaN
WORLDWIDES[:] = np.NaN
OPENING[:] = np.NaN

def get_all_features_mojo(url,x):
    soup = getAndParseURL(url)
    budget_html_item = soup.findAll("div",{"class":"a-section a-spacing-none"})

    for i in budget_html_item:
    #budget = i.findAll("span")
        txt = i.text
        #print(txt)
        if not txt.find("Budget"):
            BUDGETS[x] = (int(txt.replace("Budget$","").replace(",","")))

        if not txt.find("Domestic Distributor"):
            DOMESTIC_DISTRIBUTORS[x] = (txt.replace("Domestic Distributor","").replace("See full company information","").replace("\n",""))

        if not txt.find("MPAA"):
            MPAAS[x] = (txt.replace("MPAA",""))

        if not txt.find("Genres"):
            GENRES[x] = (txt.replace("Genres","").split())

        if not txt.find("Domestic Opening"):
            OPENING[x] = ( int(txt.replace("Domestic Opening$","").replace(",","")))

        if txt.find("Worldwide") == 88:
            WORLDWIDES[x] = (int(txt.replace("Worldwide","").replace("\n", "" ).replace("$","").replace(",","")))
 
```

```python
for i in range(len(url)):
    get_all_features_mojo(url[i], i)
```

```python
dt = pd.DataFrame({"Title":TT_list, "Worldwide":WORLDWIDES, "Budget":BUDGETS, "MPAA":MPAAS, "Domestic Distributor":DOMESTIC_DISTRIBUTORS,"Domestic Opening":OPENING,"Genres":GENRES})
```
